## JSON (JavaScript Object Notation)

JSON is a human-readable, lightweight format that is commonly used on the internet and in APIs. An example of a Knowledge Graph in a JSON format would querying the Google Knowledge Graph.

### Querying the Google Knowledge Graph

https://developers.google.com/knowledge-graph/

Note that querying the knowledge graph requires a Google Dev API key stored in `api_key.txt`. The result from querying "Palm Oil" can be seen below.

````python
"""Example of Python client calling Knowledge Graph Search API."""

api_key = open('api_key.txt').read()
query = 'Palm Oil'
service_url = 'https://kgsearch.googleapis.com/v1/entities:search'
params = {
    'query': query,
    'limit': 10,
    'indent': True,
    'key': api_key,
}
url = service_url + '?' + urllib.parse.urlencode(params)
response = json.loads(urllib.request.urlopen(url).read())
pprint(response)
for element in response['itemListElement']:
    print(element['result']['name'] + ' (' + str(element['resultScore']) + ')')
````

## RDF (Resource Description Framework)

The Resource Description Framework is an XML-based language to describe resources. Please note: RDFs use URIs (Uniform Resource Identifiers) so rather than seeing something like "tree bark" in an RDF you will see something like "www.example.org/ontology/plant/tree/#bark". Additionally, there are two kinds of notations for RDFs, the subject-predicate-object notation, called **N3** notation, which is more intuitive and the **serialized** format which is all of the triples aggregated and nested in a huge mess. To give an example, consider the following three sentences:
* Buddy Belden owns a business.
* The business has a Web site accessible at http://www.c2i2.com/~budstv.
* Buddy is the father of Lynne.

The N3 or subject-predicate-object notation is:
* <#Buddy> <#owns> <#business>.
* <#business> <#has-website> \<http://www.c2i2.com/~budstv >.
* <#Buddy> <#father-of> <#Lynne>

## Takeaways

We want to choose a data format that:

* Is **flexible** and works with a variety of tools/pipelines.
* Allows for complex types such as **temporal** and **spatial** attributes.
* Allows for **source attribution** either directly or in metadata

### Recommendation:

I recommend we use CSVs, with separate CSVs for each kind of entity and relation. In the headers we include information about the type and a separate metadata file for the source of each piece of information.

* I recommend seperate files for each entity and relationship type to allow for different schema for each entity and relationship.
* I recommend including types to make processing easier, to make our files more human friendly, and because some packages (Neo4j) want them.
* I recommend CSV because I think its flexible, human-readable, and easy to use in Neo4j and Networkx

This would look like the following:

`nodes_companies.csv:`

| CompanyID:UUID | OrbisName:string | OrbisHeadquarters:point({x, y, crs: 'wgs-84'}) |
|----------------|------------------|------------------------------------------------|
| 000001         | Wilmar International| ( 100, 120 ) |

`nodes_directors.csv`

| DirectorID:UUID | OrbisFName:string     | OrbisLName:string     |
|-----------------|-----------------------|-----------------------|
| 001001          | Alex                  | Michels               |

`edges_companydirectors.csv`

| EdgeUUID:UUID | SourceNode:UUID | TargetNode:UUID | Role:string       |
|---------------|-----------------|-----------------|-------------------|
| 100001        | 001001          | 0000001         | Director          |

`metadata.txt`

